```python
import random                   # Usaremos el método choice de random para escoger preguntas aleatoriamente

#Creamos un banco de preguntas de Geografía
banco_preguntas = {'¿Cuál es el país más pequeño del mundo?' : 'Vaticano', 
                    '¿Qué país tiene más habitantes?' : 'China', 
                    '¿Qué país es el más grande del mundo?' : 'Rusia',
                    '¿Cuál es la montaña más alta del mundo?' : 'Everest', 
                    '¿Cómo se llama el río que atraviesa la selva tropical de Brasil?' : 'Amazonas',
                    '¿Cuál es el río más largo del mundo?' : 'Nilo',
                    "¿Cual es la capital de Italia?" : "Roma", 
                    "¿En qué ciudad se encuentra la Sagrada Familia?" : "Barcelona", 
                    "¿Que país tiene la mayor población de África?" : "Nigeria",
                    "¿Cuál es la ciudad más poblada del mundo?" : "Tokio",
                    "¿Cuál es la montaña más alta de España?": "Teide", 
                    "¿Qué río pasa por París?" : "Sena", 
                    "¿Cuál es la capital de Suiza?" : "Berna",
                    "El continente más poblado es..." : "Asia"
                  }

#Damos la bienvenida el jugador, explicamos instrucciones para jugar
print("<---Bienvenido al Geo-Trivial!--->")
print(("+" + " ")*50)
print("Acierta 5 preguntas y serás coronado: REY DEL TRIVIAL.\n Falla 3 respuestas y recibirás un suspenso en Geografía")
print(("+" + " ")*50)
print(">>>COMENZAMOS!!! (>‿◠)✌")
print(("+" + " ")*50)
print("----------------------------------------")

salir_del_programa = False     # Este bucle permitirá al usuario seguir encadenando partidas después de ganar o perder. 

while (not salir_del_programa):
     
    # Iniciamos el numero de vidas del jugador. Perderá si llegan a 0.
    vidas = 3

    #Iniciamos el numero de aciertos del jugador. Ganará si llegan a 5.
    aciertos = 0

    seguir_jugando = True

    while seguir_jugando :
     
        pregunta, respuesta = random.choice(list(banco_preguntas.items()))         
        respuesta = respuesta.lower()  # Aplicamos el método lower() para evitar errores uso incorrecto de mayúsculas 
        print(f"Atención, pregunta: {pregunta}")
        respuesta_usuario = input(f"Por favor, ingresa tu respuesta").lower() 
        if respuesta_usuario == respuesta:
            aciertos += 1              
            print("¡Respuesta correcta!")
        else:
            vidas -= 1
            print("Respuesta incorrecta :(")

        if vidas == 0 or aciertos == 5:     # El juego sigue haciendo preguntas hasta que el usuario gane o pierda
            seguir_jugando = False
        else: 
            seguir_jugando = True
            print(f"RECUENTO: Llevas {aciertos} aciertos. Te quedan {vidas} vidas\n")
              
    # El juego sale del bucle por dos motivos: o el usuario ha gastado sus 3 vidas o ha acertado 5 respuestas

    if aciertos == 5:
            print("¡ENHORABUENA! Te has pasado el juego. Te coronamos: REY DEL TRIVIAL.")
    else:
            print("Ohhhhh, has perdido todas tus vidas. Te regalamos un suspenso en Geografía: '0' ")   

    otra_partida = input("¿Quieres echar otra partida? Pulsa 'J' para JUGAR DE NUEVO o 'S' para SALIR").lower()
    
    if otra_partida == 'j':
        salir_del_programa = False
        print("¡VAMOS A JUGAR OTRA RONDA!")
    else:
        salir_del_programa = True
        print("¡Hasta otra!")
```
