 # Ahorcado

- Reglas del juego del "Ahorcado":

    - Un jugador elige una palabra secreta y dibuja un espacio para cada letra de la palabra.

    - El otro jugador intenta adivinar letras para completar la palabra.

    - Si el jugador adivinador adivina una letra correctamente, se revela en su lugar correspondiente.

    - Si el jugador adivinador adivina incorrectamente, se dibuja una parte del cuerpo en la horca.

    - El objetivo del jugador adivinador es adivinar la palabra antes de que se dibuje el dibujo completo en la horca.


- Pistas para las alumnas:

    1. Empieza por seleccionar una palabra secreta para que el jugador adivinador intente adivinar. Puedes crear una lista de palabras.

    2. Inicializa el tablero de juego con guiones bajos ("_") para representar las letras no adivinadas.

    3. Utiliza un bucle `while` para permitir que el jugador adivinador haga múltiples intentos hasta que adivine la palabra o se agoten los intentos.

    4. Verifica si la letra adivinada por el jugador está en la palabra secreta y actualicen el tablero en consecuencia.

    5. Lleva un seguimiento de las letras adivinadas y los intentos fallidos.

    6. Determina las condiciones de victoria y derrota del juego, y salgan del bucle cuando se cumplan estas condiciones.

```python
import random

# Saludamos al jugador y le damos la bienvenida.
print("*********************************************************************")
print("                   ¡Bienvenid@ al juego del Ahorcado!")
print("               Un juego de: TRITURADORAS DE CÓDIGO, S.A.")
print("*********************************************************************\n")


print("               pulsa ESPACIO para leer las instrucciones\n")
input(" ")

print("\n ---------------------------------------------------------------------\n")

# Recordamos las instrucciones.
print("          Antes de comenzar, recordemos las instrucciones del juego:\n")

print("El objetivo del juego es adivinar una palabra en el menos número de intentos posibles: \n"
      "- Primero se deben marcar tantas rayas como letras tenga la palabra a adivinar.\n"
      "- Se dibuja una horca y el jugador dirá letras para adivinar la palabra.\n"
      "- Si la letra dicha por el jugador es acertada, se escribirá en la raya de la posición correspondiente.\n"
      "- Si la letra dicha por el jugador no es acertada, se dibujará una parte del monigote.\n"
      "- Hay que conseguir adivinar la palabra antes de que se complete el cuerpo...\n\n"

      "                            ¡o terminarás AHORCAD@!\n")

print("               pulsa ESPACIO para comenzar a jugar.\n")
input(" ")

print("----------------------¿PREPARAD@?----------------------\n")
print("------------------------¿LIST@?----------------------\n")
print("------------------¡COMIENZA EL JUEGO!----------------------\n")


# Inicializamos contadores y variables del juego.
vidas = 6 
letras_adivinadas = set() # Creamos una variable de tipo set para asegurarnos de que no se repitan las letras dichas por el jugador.

# La maquina elige aleatoriamente la palabra secreta de entre una lista de palabras que hemos creado. 
lista_palabras = ["mar", "gato", "barco", "cielo", "pato", "erizo"]
palabra_secreta = random.choice(lista_palabras)
set_letras_palabra_secreta = set(palabra_secreta)

# Establecemos las variables que usaremos 
lista_palabra_usuario = list("_"*len(palabra_secreta))
palabra_oculta = " ".join(lista_palabra_usuario)
set_letras_acertadas = set()
set_letras_falladas = set()

# Mostramos los espacios de las letras que el usuario tiene que adivinar con guiones bajos en pantalla.
print(f"Acabo de pensar una palabra de {len(palabra_secreta)} letras.")    
print (f"¿Serás capaz de acertarla, adivinando sus letras?\n")
print({palabra_oculta})

# El programa se desarrolla en un bucle que se ejecuta mientras el usuario tenga vidas y no haya adivinado la palabra.
seguir_jugando = vidas > 0 and set_letras_acertadas != set_letras_palabra_secreta 

while (seguir_jugando):

    letra = input("Introduce una letra").lower() # Pedimos al jugador que introduzca una letra

    if letra in palabra_secreta: 
        print("Correcto, ¡tu letra está en la palabra!")
        posicion_letra_acertada = palabra_secreta.index(letra) # Averiguamos la posición donde se encuentra la letra acertada.
        lista_palabra_usuario[posicion_letra_acertada]= letra # Sustituimos el guión por la letra acertada en la lista.
        palabra_oculta = " ".join(lista_palabra_usuario)    # Actualizamos la palabra oculta con la nueva letra.
        set_letras_acertadas.add(letra)

    else:
        set_letras_falladas.add(letra)
        vidas -= 1
        print(f"Has fallado. Pierdes una vida. Te quedan {vidas} vidas")
    
    # Informamos de las vidas restantes y letras falladas 

    print(f"Letras falladas: {set_letras_falladas}")
    seguir_jugando = vidas > 0 and set_letras_acertadas !=set_letras_palabra_secreta 
    print(palabra_oculta)
    

# Establecemos las condiciones que determinan la salida del bucle while: que el usuario acierte la palabra o pierda todas sus vidas.

if vidas == 0:
    print(f"Ohhhhh, has perdido todas tus vidas. La palabra secreta era: {palabra_secreta}")

else:
    print (f"¡ENHORABUENA! Acertaste la palabra secreta: {palabra_secreta}")
```
