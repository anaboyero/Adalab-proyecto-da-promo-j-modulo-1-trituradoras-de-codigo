```python
#Saludar al usuario y darle la bienvenida

print("---¡Bienvenido/a al juego de Trivia!---")
print("Veamos tu nivel de geografia")
print("Instrucciones: responde con una única palabra")
print("----------------------------------------")

# preparar un contador de vidas

vidas = 3

# contador de respuestas acertadas

aciertos = 0

# Creamos el diccionario con la preguntas

banco_preguntas = {'¿Cuál es el país más pequeño del mundo?' : 'Vaticano', 
                        '¿Qué país tiene más habitantes?' : 'China', 
                         '¿Qué país es el más grande del mundo?' : 'Rusia',
                         '¿Cuál es la montaña más alta del mundo?' : 'Everest', 
                         '¿Cómo se llama el río que atraviesa la selva tropical de Brasil?' : 'Amazonas',
                         '¿Cuál es el río más largo del mundo?' : 'Nilo',
                         '¿Cuál es el lugar más frío de la Tierra?' : 'Antártida',
                         "¿Cual es la capital de Australia?" : "Canberra", 
                         "¿Que pais tiene la mayor cantidad de fronteras con otros paises?" : "China", 
                         "¿Que pais tiene la mayor cantidad de islas en el mundo?" : "Suecia", 
                         "¿En que ciudad se encuentra la Sagrada Familia?" : "Barcelona", 
                         "¿Es Biden el presidente de E.E.U.U?" :  "No", 
                         "¿Que país africano tiene la mayor población?" : "Nigeria",

                  }


seguir_jugando = True

while seguir_jugando == True :

# recorre el diccionario 

    for pregunta, respuesta in  banco_preguntas.items(): 
        print(f"Empieza el bucle for. La pregunta es {pregunta} y la respuesta es {respuesta}")

        respuesta_usuario = input(f"{pregunta}")

        if respuesta_usuario != respuesta:
            print("Has fallado")
            vidas -= 1

        else:
            aciertos += 1
            print("Has acertado")
        
        if aciertos == 5:
            print("¡Ganaste! Has acertado 5 respuestas.")
            seguir_jugando = False
            break

        if vidas == 0:
            print("Lo siento, te has quedado sin vidas")
            seguir_jugando = False
            break

        print(f"Te quedan {vidas} vidas y llevas {aciertos} aciertos")
        


```
