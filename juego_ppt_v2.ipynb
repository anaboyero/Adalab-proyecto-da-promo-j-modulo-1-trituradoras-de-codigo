```python
import random
#Saludamos al jugador y le damos la bienvenida

print("            ¡Bienvenido al juego piedra, papel o tijera!\n")

print("               pulsa ESPACIO para comenzar a jugar")
input(" ")

print("\n -----------------------------------------------------------------\n")

print("               pulsa ESPACIO para leer las instrucciones\n")
input(" ")

# Recordamos las instrucciones
print("          Antes de comenzar, recordemos las instrucciones del juego:\n")

print("El objetivo del juego es vencer al oponente seleccionando el arma que gana según las siguientes reglas: \n"
      
      "- El partida consta de 3 rondas.\n"
      "- Puedes elegir un arma distinta por cada ronda.\n"
      "- La piedra aplasta a la tijera.\n"
      "- La tijera corta el papel.\n"
      "- El papel envuelve a la piedra.\n"
      "- En caso de que ambos jugadores elijan la misma arma, se produciría un empate y se juega otra vez.\n")

print("----------------------¿PREPARADOS?----------------------\n")
print("------------------------¿LISTOS?----------------------\n")
print("------------------¡COMIENZA EL JUEGO!----------------------\n")


# Comparamos ambas tiradas introduciendo un bucle while que incluya las tres condiciones
contador_rondas = 0

while contador_rondas < 3:
    arma = input("Elije tu arma para jugar") # Pedimos al usuario que elija su arma

    if arma not in opciones_armas:
        print(f"Has elegido {arma}, esta opción no es válida, elige de nuevo.")
        continue

    if arma in opciones_armas and contador_rondas == 0:
        print("\nRONDA 1")

    if arma in opciones_armas and contador_rondas == 1:
        print("\nRONDA 2")
    
    if arma in opciones_armas and contador_rondas == 2:
        print("\nRONDA 3")

    print(f"Has elegido {arma}") # Mostramos al jugador la confirmacion del arma elegida
    

    opciones_armas = ["piedra", "papel", "tijeras"]
    opcion_aleatoria = random.randint(0,2) 
    arma_maquina = opciones_armas[opcion_aleatoria] # La maquina escoge su arma de manera aleatoria
    print(f"El ordenador ha elegido {arma_maquina}") # Mostramos al jugador
    
    if arma == "piedra":
        if arma_maquina == "piedra":
            print("¡Has empatado!")
        if arma_maquina == "papel":
            print("¡Has perdido!")
        if arma_maquina == "tijera":
            print("¡Has ganado!")
            
    elif arma == "papel":
        if arma_maquina == "piedra":
            print("¡Has ganado!")
        if arma_maquina == "papel":
            print("¡Has empatado!")
        if arma_maquina == "tijera":
            print("¡Has perdido!")

    else:
        if arma_maquina == "piedra":
            print("¡Has perdido!")
        if arma_maquina == "papel":
            print("¡Has ganado!")
        if arma_maquina == "tijera":
            print("¡Has empatado!")
    
    contador_rondas +=1

print("\n----------------------SE ACABÓ LA PARTIDA----------------------.")
```
