```python
import random

#Damos la bienvenida el jugador, explicamos instrucciones para jugar
print("<---Bienvenido al juego de Trivial sobre Geografía!--->")
print(("+" + " ")*50)
print("Sigue las siguientes instrucciones y sera el vencedor:\n\n1. Solo tendras 3 intentos para responder correctamente cada pregunta\n2. Si respondes 3 preguntas correctamente ganaras el juego.\n3. Cada pregunta solo  se responde correctamente con una palabra\n" )
print(("+" + " ")*50)
print(">>>COMENZAMOS!!! (>‿◠)✌")
print(("+" + " ")*50)
print("----------------------------------------")


#Indicamos e iniciamos el numero de equivocaciones que tiene el jugador
vidas = 0

#Indicamos e iniciamos el numero de aciertos que tiene el jugador
aciertos = 0

#Creamos lista para guardar las preguntas que responde
#preguntas_respondidas = []

#Indicamos la cantidad de veces que se puede equivocar
max_intentos = 3

#Indicamos la cantidad de veces que puede acertar para ganar el juego
max_aciertos = 5

#Banco de preguntas
banco_preguntas = {'¿Cuál es el país más pequeño del mundo?' : 'vaticano', 
                    '¿Qué país tiene más habitantes?' : 'china', 
                    '¿Qué país es el más grande del mundo?' : 'rusia',
                    '¿Cuál es la montaña más alta del mundo?' : 'everest', 
                    '¿Cómo se llama el río que atraviesa la selva tropical de Brasil?' : 'amazonas',
                    '¿Cuál es el río más largo del mundo?' : 'Nnilo',
                    '¿Cuál es el lugar más frío de la Tierra?' : 'antartida',
                    "¿Cual es la capital de Australia?" : "canberra", 
                    "¿Que pais tiene la mayor cantidad de fronteras con otros paises?" : "china", 
                    "¿Que pais tiene la mayor cantidad de islas en el mundo?" : "suecia", 
                    "¿En que ciudad se encuentra la Sagrada Familia?" : "barcelona", 
                    "¿Es Biden el presidente de E.E.U.U?" :  "no", 
                    "¿Que país africano tiene la mayor población?" : "nigeria",

                  }


seguir_jugando = True


while vidas < max_intentos and aciertos < max_aciertos:
     
    pregunta, respuesta = random.choice(list(banco_preguntas.items()))
    intentos_respuesta = 0

    while intentos_respuesta < 3:
         
        #respuesta_usuario = input(f"{pregunta} Por favor ingresa tu respuesta")
        print(f"La pregunta es:{pregunta}")
        respuesta_usuario = input(f"Por favor ingresa tu respuesta")
          
        if respuesta_usuario == respuesta:
            #Se suma un acierto
            aciertos +=1
            print(f"Respuesta correcta!!.Te quedan {max_intentos} vidas y llevas {aciertos} aciertos")
            seguir_jugando = False
              
        else:
            print(f"Lo siento, la respuesta no es correcta. La respuesta correcta era '{respuesta}'.\nNo te preocupes vamos por la siguiente pregunta\n")
            vidas += 1

    if intentos_respuesta == 3:
        intentos_respuesta +=1

if aciertos == max_aciertos:
        print("¡Ganaste! Eres todo un experto en geografia")

else:
        print("Lo siento, te has quedado sin intentos")   
      
        
```
